```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json
```

```python
driver = webdriver.Chrome()
```

```python
cat_dict = {'t-shirts':[],'polo-t-shirts':[], 'ut-graphic-tees':[], "sweatshirts-and-hoodies":[]}
```

```python
for cat, shirt_list in cat_dict.items():
    driver.get(f'https://www.uniqlo.com/in/en/men/tops/{cat}')
    
    
    try:
        old_height = driver.execute_script('return document.body.scrollHeight')

        while True:
            for i in driver.find_elements(by=By.CLASS_NAME , value='text'):
                i.click()
                time.sleep(1)
                break


            new_height = driver.execute_script('return document.body.scrollHeight')

            if new_height == old_height:
                break

            old_height = new_height
    except:
        pass
    
    for i in driver.find_elements(by=By.XPATH, value='//a[@data-category="ProductCollection-grid"]'):
        shirt_list.append(i.get_attribute("href"))
    
```

```python
counter = 1

product_id = []
product_name = []
category = []
price = []
og_price = []
color = []
size = []
rating_5 = []
rating_4 = []
rating_3 = []
rating_2 = []
rating_1 = []
stock_availability = []




def extract_text(element):
    try:
        return element.text
    except:
        return np.nan

    
for cat, links in cat_dict.items():
for link in links: 
    driver.get(link)

    try:
        color_n_size_elements = driver.find_elements(By.XPATH, '//div[@class="fr-chip-wrapper"]')
        rating_elements = driver.find_elements(By.XPATH, '//div[@class="fr-review-stars-with-bar-count"]')
        product_name_element = driver.find_element(By.XPATH, '//span[@class="title fr-no-uppercase"]')
        prize_elements = driver.find_elements(By.XPATH, '//span[@class="fr-price-currency xN1x05O0jJUkQaFHARtFH"]')
        radio_buttons = driver.find_elements(By.XPATH, '//input[@type="radio"][@name="product-size-picker"]')
        size_elements = driver.find_elements(By.XPATH, '//span[@class="fr-chip-text"]')

        for color_size in color_n_size_elements:
            color_name = color_size.get_attribute("data-test")
            if color_name not in ["S", "M", "L", "XL", "XXL", "3XL"]:
                color_size.click()

                for num, size_ in enumerate(size_elements):
                    is_disabled = radio_buttons[num].get_attribute("disabled")
                    stock_status = "Out of Stock" if is_disabled else "Available"

                    product_id.append(f"UNI{counter}")
                    product_name.append(extract_text(product_name_element))
                    category.append(cat)
                    price.append(extract_text(prize_elements[1]) if len(prize_elements) > 1 else np.nan)
                    og_price.append(extract_text(prize_elements[0]) if len(prize_elements) > 0 else np.nan)
                    color.append(color_name)
                    size.append(size_.text)

                    for i, rating in enumerate([rating_5, rating_4, rating_3, rating_2, rating_1]):
                        rating.append(extract_text(rating_elements[i]) if len(rating_elements) > i else np.nan)

                    stock_availability.append(stock_status)

                    counter += 1
            else:
                break
    except Exception as e:
        pass

    time.sleep(2)
```

```python
raw_df = pd.DataFrame({'product_id' : product_id,
'product_name' : product_name,
'category' : category,
'price' : price,
'og_price' : og_price,
'size' :size,
'rating_5' : rating_5,
'rating_4' : rating_4,
'rating_3' : rating_3,
'rating_2' : rating_2,
'rating_1' : rating_1,
'stock_availability':stock_availability})
    
```
