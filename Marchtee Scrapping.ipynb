```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json
```

```python
base_url = "https://www.marchtee.com"
driver1 = webdriver.Chrome()


def get_data(category_value):
    
    url = f"https://www.marchtee.com/{category_value}/men/navy"
    driver1.get(url)

    urls=[]
    for i in (driver1.find_elements(by = By.XPATH,value='//a[@data-url]')):
        urls.append(i.get_attribute('data-url'))
        
        
    product_id = []
    category = []
    size = []
    dis_price = []
    og_price=[]
    color = []
    multiple_pack = []
    availibility_of_stock = []
    return_time = []
    
    counter = 0

    for j,i in enumerate(urls[1:]):
        driver1.get(f'{base_url}{i}')  
        time.sleep(3)
        inner_counter
        for i in driver1.find_elements(by = By.XPATH, value = '//a[@data-price]'):

            h2_heading = driver1.find_elements(by=By.XPATH, value = "//h2[@class='serif para feature']")[0].text
            h2_heading_list = h2_heading.split('\n') 

            product_status = driver1.find_elements(by=By.XPATH, value = "//span[@id='productStatus']")[0].text

            category.append(driver1.find_elements(by = By.XPATH, value = f'//a[@data-url="/{category_value}#hero"]')[0].text) 
            
            product_id.append(f"MT{counter}")
            size.append(i.text)
            dis_price.append(i.get_attribute("data-price"))
            og_price.append(i.get_attribute("data-original-price"))
            return_time.append(driver1.find_elements(by=By.XPATH, value = "//span[@id='returnsInfo']")[0].text)


            if h2_heading_list[0] == 'Pack of 3':
                multiple_pack.append(1)
                color.append(h2_heading_list[1])
            else:
                multiple_pack.append(0)
                color.append(h2_heading_list[0])


            if product_status == 'Out of Stock':
                availibility_of_stock.append('Out of Stock')
            else:
                availibility_of_stock.append("Available")
        
        
            inner_counter+=1    
        
                
    return (pd.DataFrame(data= {'product_id':product_id,'category':category, 'size':size, 'price':dis_price, 
                                'og_price':og_price, 'color':color, 'multiple_pack':multiple_pack, 
                                'availibility_of_stock':availibility_of_stock, 'return_time':return_time, }))

            



category_value = ["tee", "polo", "v"]

raw_df = pd.DataFrame(columns = ['product_id','category', 'size', 'price', 'og_price', 'color',
                         'multiple_pack', 'availibility_of_stock', 'return_time',])


for i in category_value:
    returned_df = get_data(i)
    
    raw_df = pd.concat([raw_df, returned_df])
```
